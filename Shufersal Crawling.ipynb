since that the items in the website are loaded by scrolling down instead of switching pages ill have to use a function to scroll down to the end of the page before i start finding information that i need
at the end of the proccess the function will get all the elements( all of the items in page) of the class 
and return the list of the items

```python
def LoadAllPage(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        time.sleep(3)
        
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height    
    
    items=driver.find_elements(By.CLASS_NAME,'miglog-prod.miglog-sellingmethod-by_unit')
    return items
```


InfoFromItem function:

ill use a function that will work everytime item's pop-up opened 
at this stage im able to take the data from the item
but before that, basic conditions need to be checked:
if the function was called but the popup didnt open, the function will return None
if the pop up opened but in this item there is no nutrition list, return None

once all the conditions are met the function will save the item's name
then it will find the class that has the element of the nutrtion list and save the elements in a list
next, it will go through the elements and copy the element name into a list and also the number it provides into another list, these two lists are: title and amount

after taking all the nutrition list the function will try to find the symbols
in order to do that, ill use the try-except method, so the if i cant find the XPATH it means there is no symbol
the way to identify the symbol is by the end of the XPATH for example: red.symbol.classfication.320
320 means is sodium, 321 means sugar, 322 means fat and 324 means the green symbol

after finding all the information that i need the function will put all of the information in one dataframe
when the titles of the nutrition list will be the columns, in addition, it will create 4 more columns for the symbols

the function will return a dataframe for 1 item
 

```python
def InfoFromItem(driver):
    #print('success')
    titlelist=[]
    amountlist=[]
    flag=0
    
    try:
        workable=driver.find_element(By.CLASS_NAME,"modal.fade.productModal.centerModal.bodyModalAppend.in")
    except:
        print("popup not found")
        return None 
    
    try:
        workable=driver.find_element(By.CLASS_NAME,"title.nutritionListTitle")
    except:
        print("no nutritions found")
        return None 
    
   
    
    #print("getting nutrition list")
    nutritions=driver.find_elements(By.CLASS_NAME,'nutritionItem')
      
    #print("success")
    itemname=driver.find_element(By.XPATH,'//*[@id="modalTitle"]').text
    try:
       # print("getting nutrition info")
        
        for item in nutritions: 
            
            title= item.find_element(By.CLASS_NAME,'text').text
            amount=item.find_element(By.CLASS_NAME,'number.tooltip-js').text
            
            try:
                symbol=item.find_element(By.XPATH,"//img[@alt='red.symbol.classfication.320']")# FIND SODIUM MARK
                sodium=1
            except:
                sodium=0
            
            try:
                symbol=item.find_element(By.XPATH,"//img[@alt='red.symbol.classfication.321']")#FIND SUGAR MARK
                sugar=1
            except:
                sugar=0
       
            try:
                symbol=item.find_element(By.XPATH,"//img[@alt='red.symbol.classfication.322']")#FIND FAT MARK
                fat=1
            except:
                fat=0
            
            try:
                symbol=item.find_element(By.XPATH,"//img[@alt='red.symbol.classfication.324']")#FIND GREEN MARK
                green=1
            except:
                green=0
        
            titlelist.append(title)
            amountlist.append(amount)
    except:
        print("error in the for loop")
    #print("success")
   # print("creating dataframe")
    try:
        df = pd.DataFrame([amountlist],
                  columns=titlelist)
        df = df.loc[:,~df.columns.duplicated()]
    except:
        print("error while creating dataframe")
    df['שם המוצר']=itemname
    df['סוכר בכמות גבוהה']=sugar
    df['נתרן בכמות גבוהה']=sodium
    df['שומן רווי בכמות גבוהה']=fat
    df['הסימון הירוק']=green
   # print("success")
    return df 
```
